```python
# Grouping by Election Date and County Name
verification_group = result_df.groupby(['Election Date', 'County Name'])

# Initialize a list to collect any inconsistencies
inconsistencies = []

# Check consistency of candidate names for each group
for (election_date, county_name), group in verification_group:
    # Extract unique candidates for each party in the group
    rep_candidates = group['REP Candidate'].unique()
    dem_candidates = group['DEM Candidate'].unique()
    other_candidates = group['Other Candidate'].unique()

    # Check if there is more than one unique candidate name for any party
    if len(rep_candidates) > 1:
        inconsistencies.append({
            'Election Date': election_date,
            'County Name': county_name,
            'Party': 'REP',
            'Candidates': rep_candidates.tolist()
        })
    if len(dem_candidates) > 1:
        inconsistencies.append({
            'Election Date': election_date,
            'County Name': county_name,
            'Party': 'DEM',
            'Candidates': dem_candidates.tolist()
        })
    if len(other_candidates) > 1:
        inconsistencies.append({
            'Election Date': election_date,
            'County Name': county_name,
            'Party': 'OTHER',
            'Candidates': other_candidates.tolist()
        })

# Convert inconsistencies to a DataFrame for easier review
inconsistencies_df = pd.DataFrame(inconsistencies)
```

```python
# Grouping by Election Date and County Name
county_grouped = result_df.groupby(['Election Date', 'County Code', 'County Name'])

# Initialize an empty list to store the county-wise aggregated data
county_aggregated_rows = []

# Iterate over each group
for (election_date, county_code, county_name), group in county_grouped:
    # Aggregating data
    total_registered_voters = group['Total Registered Voters'].sum()
    rep_votes = group['REP Votes'].sum()
    dem_votes = group['DEM Votes'].sum()
    other_votes = group['Other Votes'].sum()
    total_votes = group['Total Votes'].sum()

    # Extracting candidate names (assumes one unique name per party per county)
    rep_candidate = group['REP Candidate'].unique()
    dem_candidate = group['DEM Candidate'].unique()
    other_candidate = group['Other Candidate'].unique()

    # Resolving to single candidate names or empty string
    rep_candidate = rep_candidate[0] if len(rep_candidate) == 1 else ''
    dem_candidate = dem_candidate[0] if len(dem_candidate) == 1 else ''
    other_candidate = other_candidate[0] if len(other_candidate) == 1 else ''

    # Create a single row for the county
    county_aggregated_rows.append({
        'County Code': county_code,
        'County Name': county_name,
        'Election Date': election_date,
        'Total Registered Voters': total_registered_voters,
        'REP Candidate': rep_candidate,
        'DEM Candidate': dem_candidate,
        'Other Candidate': other_candidate,
        'REP Votes': rep_votes,
        'DEM Votes': dem_votes,
        'Other Votes': other_votes,
        'Total Votes': total_votes
    })

# Create a new DataFrame from the aggregated rows
county_result_df = pd.DataFrame(county_aggregated_rows)
```

```python
county_result_df.head()
```

```python
# Grouping by Election Date
election_grouped = county_result_df.groupby('Election Date')

# Initialize an empty list to store aggregated election data
election_aggregated_rows = []

# Iterate over each group
for election_date, group in election_grouped:
    # Aggregating data
    total_registered_voters = group['Total Registered Voters'].sum()
    rep_votes = group['REP Votes'].sum()
    dem_votes = group['DEM Votes'].sum()
    other_votes = group['Other Votes'].sum()
    total_votes = group['Total Votes'].sum()

    # Calculating voter turnout
    voter_turnout = (total_votes / total_registered_voters) * 100 if total_registered_voters > 0 else 0

    # Determining the winner
    votes_dict = {
        'REP': rep_votes,
        'DEM': dem_votes,
        'OTHER': other_votes
    }
    won = max(votes_dict, key=votes_dict.get)

    # Append the aggregated data
    election_aggregated_rows.append({
        'Election Date': election_date,
        'Total Registered Voters': total_registered_voters,
        'REP Votes': rep_votes,
        'DEM Votes': dem_votes,
        'Other Votes': other_votes,
        'Total Votes': total_votes,
        'Voter Turnout (%)': voter_turnout,
        'WON': won
    })

# Create a new DataFrame from the aggregated rows
election_result_df = pd.DataFrame(election_aggregated_rows)

# Sort by election year (assuming Election Date is in datetime format or can be converted)
election_result_df['Election Year'] = pd.to_datetime(election_result_df['Election Date']).dt.year
election_result_df.sort_values(by='Election Year', inplace=True)
election_result_df.drop(columns=['Election Year'], inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression

# Preparing data for classification (predicting "WON")
X_classification = df[['Total Registered Voters', 'REP Votes', 'DEM Votes', 'Other Votes']]
y_classification = df['WON']

# Split data for training and testing
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_classification, y_classification, test_size=0.2, random_state=42
)

# Naive Bayes Classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train_class, y_train_class)

# Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train_class, y_train_class)

# Predictions for 2024 (extrapolated values)
future_data = pd.DataFrame({
    'Total Registered Voters': [620000],
    'REP Votes': [230000],
    'DEM Votes': [150000],
    'Other Votes': [0]
})

nb_prediction = nb_classifier.predict(future_data)
dt_prediction = dt_classifier.predict(future_data)

# Preparing data for regression (predicting "Voter Turnout (%)")
X_regression = df[['Total Registered Voters', 'REP Votes', 'DEM Votes', 'Other Votes']]
y_regression = df['Voter Turnout (%)']

# Split data for training and testing
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_regression, y_regression, test_size=0.2, random_state=42
)

# Linear Regression Model
regression_model = LinearRegression()
regression_model.fit(X_train_reg, y_train_reg)

# Prediction for voter turnout in 2024
turnout_prediction = regression_model.predict(future_data)

# Display predictions for 2024
predictions_2024 = pd.DataFrame({
    'Model': ['Naive Bayes', 'Decision Tree', 'Voter Turnout'],
    'Prediction': [nb_prediction[0], dt_prediction[0], turnout_prediction[0]]
})

print(predictions_2024)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Convert to DataFrame
df = pd.DataFrame.from_dict(adjacency_list, orient='index').reset_index()
df.columns = ['State'] + list(df.columns[1:])

# Step 2: Calculate win counts and percentages
win_counts = df.set_index('State').apply(lambda x: pd.Series(x.value_counts()), axis=1).fillna(0)
win_counts['Dem_Wins'] = win_counts.get(0, 0)
win_counts['Rep_Wins'] = win_counts.get(1, 0)

win_counts['Total_Elections'] = win_counts['Dem_Wins'] + win_counts['Rep_Wins']
win_counts['Dem_Percentage'] = (win_counts['Dem_Wins'] / win_counts['Total_Elections']).fillna(0) * 100
win_counts['Rep_Percentage'] = (win_counts['Rep_Wins'] / win_counts['Total_Elections']).fillna(0) * 100

# Calculate Variance of Win Percentages
win_variance = df.set_index('State').var(axis=1)
win_counts['Variance'] = win_variance

# Define swing state criteria (for example, variance > 0.25)
swing_state_threshold = 0.25  # This threshold can be adjusted
win_counts['Is_Swing_State'] = win_counts['Variance'] > swing_state_threshold

# Reset index for plotting
win_counts.reset_index(inplace=True)

# Step 3: Visualization of Swing States
plt.figure(figsize=(12, 6))
sns.barplot(data=win_counts, x='State', y='Variance', hue='Is_Swing_State', palette={True: 'orange', False: 'grey'})
plt.axhline(y=swing_state_threshold, color='red', linestyle='--', label='Swing State Threshold')
plt.title('Variance of Win Percentages by State')
plt.xlabel('States')
plt.ylabel('Variance')
plt.xticks(rotation=45)
plt.legend(title='Swing State', labels=['Swing State', 'Not a Swing State'])
plt.tight_layout()
plt.show()

# Print the swing states
swing_states = win_counts[win_counts['Is_Swing_State']][['State', 'Variance']]
print("Swing States based on variance:\n", swing_states)
```

```python
import geopandas as gpd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def calculate_similarity(election_data, neighbors):
    similarity_scores = {state: {} for state in election_data.keys()}
    # For each year in the election data
    for year in next(iter(election_data.values())).keys():
        for state, results in election_data.items():
            if year not in results:
                continue
            # Get the election result for the current state
            state_result = results[year]
            # Get neighbors' results
            neighbor_results = []
            for neighbor in neighbors.get(state, []):
                if neighbor in election_data:  # Only consider valid states
                    neighbor_results.append(election_data[neighbor].get(year, None))
            # Filter out None values for neighbors
            neighbor_results = [result for result in neighbor_results if result is not None]
            # Calculate similarity measure (e.g., percentage of neighbors with the same outcome)
            if neighbor_results:
                similarity = sum(1 for result in neighbor_results if result == state_result) / len(neighbor_results)
            else:
                similarity = np.nan  # No neighbors to compare
            similarity_scores[state][year] = similarity
    return similarity_scores

# Calculate similarity scores
similarity_scores = calculate_similarity(adjacency_list, neighbors)

# Convert similarity scores to a DataFrame
similarity_df = pd.DataFrame(similarity_scores).T

# Calculate the average similarity score across years for each state
similarity_df['average_similarity'] = similarity_df.mean(axis=1)
# Ensure 'average_similarity' column has no NaN values and is of numeric type
similarity_df.reset_index(inplace = True)
df = similarity_df[['index', 'average_similarity']]
df['average_similarity'] = pd.to_numeric(df['average_similarity'].fillna(0))

# Load USA map shapefile
shapefile_path = 'Shapes/ne_110m_admin_1_states_provinces.shp'
usa_map = gpd.read_file(shapefile_path)
df['index'] = df['index'].str.title()
df.rename(columns={'index': 'name'}, inplace=True)
# Merge similarity data with the GeoDataFrame and check the result
usa_map = usa_map.merge(df, on='name', how='left')

# Plot the map with 'average_similarity' as the heat map column
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
usa_map.boundary.plot(ax=ax, linewidth=1)  # Plot state boundaries
usa_map.plot(column='average_similarity', ax=ax, legend=True,
             cmap='YlOrRd', vmin=0, vmax=1,  # Set colormap and color scale
             missing_kwds={"color": "lightgrey"},
             legend_kwds={'label': "Average Similarity Score",
                          'orientation': "horizontal"})

plt.title('USA Map with Average Similarity Scores')
plt.show()
```
